```python
import json
from sacremoses import MosesTokenizer
import numpy as np
from time import time
import random
random.seed(1989)
```

```python
reviews = json.load(open("reviews_small.json"))
```

```python
valid_ratio = 0.1
```

```python
start_time = time()
mt = MosesTokenizer(lang='en')
train_scores = []
valid_scores = []
# around 15mn
with open("train.text.en", "w") as f:
    with open("valid.text.en", "w") as g:
        for i, item in enumerate(reviews):
            if "stars" in item and "text" in item:
                if random.random() > valid_ratio:
                    f.write(" ".join(mt.tokenize(item['text'])).lower() + "\n")
                    train_scores.append(float(item['stars']))
                else:
                    g.write(" ".join(mt.tokenize(item['text'])).lower() + "\n")
                    valid_scores.append(float(item['stars']))
            if i % 1000 == 0:
                print(i)
                print(time() - start_time)
train_scores = np.array(train_scores)
np.save("train.scores", train_scores)
valid_scores = np.array(valid_scores)
np.save("valid.scores", valid_scores)
```
